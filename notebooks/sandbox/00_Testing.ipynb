```python
import geopandas as gpd
import matplotlib.pyplot as plt

# 1. Cargar el shapefile de ecorregiones
# Nota: Asegrate de que el CRS coincida con tu RASTER_DEM.tif (ej. EPSG:32718 para Per WGS84/UTM 18S)
path_ecorreg = "../data/trash/TEM_BIOLO_ECORREGIONES_FINAL.shp"
gdf_ecorreg = gpd.read_file(path_ecorreg)

# 2. Visualizacin rpida
fig, ax = plt.subplots(figsize=(10, 8))
gdf_ecorreg.plot(column='ecorregion', cmap='tab20', legend=True, 
                 legend_kwds={'title': "Ecorregiones", 'loc': 'upper left', 'bbox_to_anchor': (1, 1)},
                 ax=ax)
ax.set_title("Distribucin Espacial de Ecorregiones", fontsize=14)
plt.tight_layout()
plt.show()

# 3. Ver los datos (Atributos)
print("Columnas disponibles:", gdf_ecorreg.columns.tolist())
print(gdf_ecorreg[['ecorregion', 'geometry']].head())
```
